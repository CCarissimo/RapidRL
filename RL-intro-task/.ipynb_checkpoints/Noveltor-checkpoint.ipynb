```python
from functions import *
from agent import Qlearner
from environments import gridworld
import matplotlib.pyplot as plt
import numpy as np
from tqdm.auto import tqdm
```

```python
# Create the gridworld, and set cell values to equal rewards
print('(Creating Willemsen Gridworld)')
grid = np.ones((3, 9)) * -1
grid[1,:8] = 0
grid[1, 8] = 0.1
terminal_states = np.array([[1, 8]])
initial_state = np.array([1, 0])
blacked_state = np.array([[0, 8],[2, 8]])

fig, ax = plt.subplots()
cax = fig.add_axes([0.27, 0.75, 0.5, 0.05])
colormap = plt.get_cmap('RdYlBu')
im = ax.imshow(grid, cmap=colormap)
ax.set_xticks(np.arange(-.5, 8, 1), minor=True)
ax.set_yticks(np.arange(-.5, 3, 1), minor=True)
ax.grid(which='minor', color='w', linewidth=2)
cb = fig.colorbar(im, ax=ax, cax=cax, orientation='horizontal', label='reward')
cb.ax.xaxis.set_ticks_position('top')
cb.ax.xaxis.set_label_position('top')
cb.ax.tick_params(labelsize=10)
for i in range(len(grid[:,0])):
    for j in range(len(grid[0,:])):
        text = ax.text(j, i, grid[i, j],
                       ha="center", va="center", color="w")
plt.show()

n_episodes = 100
max_steps = 40
gamma = 0.8
alpha = 0.1
epsilon = 0.05

Learn = True
batch_size = 1
```

```python
print('(Generating Trajectories and Learning Loop)')
env = gridworld(grid, terminal_states, initial_state, blacked_state, max_steps)
agent = Qlearner(alpha, epsilon, gamma, batch_size, initial_state, s_prior=None, Learn=True)

epsilons = np.flip(np.linspace(0, 1, n_episodes))
trajectories = []

for i in tqdm(range(n_episodes)):
    state = env.reset()
    agent.reset_trajectory()
    agent.epsilon = epsilons[i]
    terminal = False

    while not terminal:

        action = agent.select_action(state)
        
        next_state, reward, terminal = env.step(action)

        agent.observe_transition(state, action, reward, next_state, terminal)

        state = next_state
    
    trajectories.append(agent.trajectory)
```

```python
print('(Calculating Returns, Abstractions, Features, Visits, Biases)')
abstraction = {}

returns_table = cumulative_table(trajectories, gamma, Qpi_sa, agent.q_table)

rho = state_dist(agent.visits)

for action in env.actions:
    if action in abstraction.keys():
        pass
    else:
        abstraction[action] = Qpi_a(action, returns_table, rho)

sa_values = []
a_values = []
for sa, v in returns_table.items():
    sa_values.append(v)
    for a, v_abs in abstraction.items():
        if a in sa:
            a_values.append(v_abs)

bias_squared = [(sa_values[i] - a_values[i]) ** 2 for i in range(len(sa_values))]

heatmap = np.copy(grid)*0
heatmapQ = np.copy(grid)*0

for k, prob in rho.items():
    heatmap[int(k[1]), int(k[3])] = prob

for state, actions in agent.q_table.items():
    heatmapQ[int(state[1]), int(state[3])] = max(qval for qval in actions.values())
```

```python
# First plot
%matplotlib notebook
print('(Plotting 1 of 1)')
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(10, 8))

ax[0, 0].scatter(sa_values, a_values)
ax[0, 0].set_title('Bias plot')
ax[0, 0].set_xlabel('state action pair value')
ax[0, 0].set_ylabel('action abstraction value')
#ax[0, 0].set_ylim((0, 40))

ax[0, 1].hist(bias_squared, cumulative=True, bins=25)
ax[0, 1].set_title('Bias squared plot')
ax[0, 1].set_xlabel(r'$(v_{sa} - v_{a})^2$')
ax[0, 1].set_ylabel('freq')
# plt.ylim((0, 40))

ax[1, 0].imshow(heatmap, cmap='binary')
ax[1, 0].set_title('gridworld visits heatmap')

ax[1, 1].imshow(heatmapQ, cmap='binary')
ax[1, 1].set_title('gridworld qvalues')

plt.tight_layout()
plt.show()
```

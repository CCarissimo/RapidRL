```python
from functions import *
from agent import Abstractor
from environments import gridworld
import matplotlib.pyplot as plt
import numpy as np
from tqdm.auto import tqdm
```

```python
# Create the gridworld, and set cell values to equal rewards
print('(Creating Willemsen Gridworld)')
grid = np.ones((3, 9)) * -1
grid[1,:8] = 0
grid[1, 8] = 0.1
grid[2, :8] = 0
terminal_state = []
for i in [0,2]:
    for j in range(8):
        terminal_state.append([i,j])
terminal_state.append([1,8])
terminal_state = np.array(terminal_state)
initial_state = np.array([1, 0])
blacked_state = np.array([[0, 8],[2, 8]])

fig, ax, im, cb = plot_gridworld(grid, terminal_state, initial_state, blacked_state)
```

```python
n_episodes = 100
max_steps = 100
gamma = 0.8
alpha = 0.1
epsilon = 0.05

Learn = True
batch_size = 1
```

## Testing a simple Abstractor over Actions in two rounds

## Reset the Environment and Agent every time

```python
env = gridworld(grid, terminal_state, initial_state, blacked_state, max_steps)
agent = Abstractor(alpha, epsilon, gamma, batch_size, initial_state, s_prior=None, learn=True)

epsilons = np.flip(np.linspace(0, 1, n_episodes))
trajectories = []
```

## First Run

```python
trajectory = run_trajectory(env, agent, 1, abstract = False)

plot_trajectory(trajectory, grid, terminal_state, initial_state, blacked_state)

print(env.terminal)
```

## Second Run

```python
trajectory = run_trajectory(env, agent, 0.2, abstract = False)

plot_trajectory(trajectory, grid, terminal_state, initial_state, blacked_state)

print(env.terminal)
```

## Third Run

```python
trajectory = run_trajectory(env, agent, epsilons[1], abstract = True)

plot_trajectory(trajectory, grid, terminal_state, initial_state, blacked_state)

print(env.terminal)
```
